# Data Science Individual Project
## Objective

Provide a data-driven solution to a problem that excites you using the tools discussed (or related to) in this course.

## Datasets
The dataset must be different with your group project final topic. 

## Tasks
For your project, you should:
- Pick an issue/problem that excites you
- Create at least three questions in the topic/issue/problem to help your solve the problem
- Select or create datasets
- Familiarize yourself with that data, if necessary:
    - data munging
    - feature engineering
- Choose proper model/method
    - Train/Fit the model by the datasets
    - Potential Methods: Classification, Regression, ...
    - Potential Tools: SciKit, TensorFlow (tf), ...
- Analyze results
- Future work
- Reference

## Rubric
|Category | Explanation| %|
|--|--|-|
|Abstract&Keyword| What the key information about the project?|5|
|Introduction|What the topic? Why was the project undertaken? What's current research/results in this topic? What was the research question？|10|
|Datasets|What is the source of the dataset? Characteristics of data? Any munging, imputation, or feature engineering?|10|
|Methodology|What materials/model/quantitative tools were used in answering the research question? What's the advantage/disadvantage of the tools? Why choose it?|20|
|Results|What answer was found to the research question; what did the study find? Any unexpected results? What's the potnetial cause/reason behind it?|20|
|Discussion|What might the answer imply and why does it matter? How does it fit in with what other researchers have found? What are the perspectives for future research? |10|
|Conclusion| What's the most important results did you find? What’s the effect of this result on the real-world?|10|
|Reference|Use the IEEE format for all references. Include [number] for citation in the report|5|
|Coding|ipynb file with clear comments in folder `codes` and data in folder `data`.|10|

Rubric based on the IMRAD:https://en.wikipedia.org/wiki/IMRAD
